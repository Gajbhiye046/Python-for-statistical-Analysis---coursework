```python
import numpy as np
import matplotlib.pyplot as plt
d1=np.loadtxt('outlier_1d.txt')
d2=np.loadtxt('outlier_2d.txt')
d3=np.loadtxt('outlier_curve.txt')
print(d1.shape,d2.shape,d3.shape)

plt.scatter(d1,np.random.normal(7,0.2,size=d1.size),s=1,alpha=0.5)
#alpha for transparency and s= dot size
plt.scatter(d2[:,0],d2[:,1])
plt.show();
plt.plot(d3[:,0],d3[:,1]);
```

How you deal with the outliers:(Manually)

```python
mean,std= np.mean(d1),np.std(d1)
z_score=np.abs((d1 - mean) / std)
threshold = 3 
good = z_score < threshold
print(f'Rejection {(~good).sum()} points')
from scipy.stats import norm
print(f'Z-score of 3 corresponds to probability of {100*2*norm.sf(threshold): 0.2f}%')
# here in calculation of probibity of z-score beyond 3 norm.sf(z-score=3) multiplied by 2 as its bell curve
visual_scatter = np.random.normal(size= d1.size)
plt.scatter(d1[good],visual_scatter[good],s=2,label= 'Good',color='#4CAF50')
plt.scatter(d1[~good],visual_scatter[~good],s=2,label = 'Bad',color='#F44336')
plt.legend();

```

```python
from scipy.stats import multivariate_normal as mn
mean,cov= np.mean(d2,axis=0),np.cov(d2.T)
#Here in numpy.cov(m),where m represents Each row of m represents a variable, and each column a single observation of all those variables thats why d2.T is m. 
good=mn(mean,cov).pdf(d2)> 0.01/100


plt.scatter(d2[good,0],d2[good,1],s=2,label= 'Good',color='#4CAF50')
plt.scatter(d2[~good,0],d2[~good,1],s=2,label = 'Bad',color='#F44336')
plt.legend();

```

```python
xs,ys= d3.T
p= np.polyfit(xs,ys,deg= 5)
ps= np.polyval(p,xs)
plt.plot(xs,ys,'.',label='Data')
plt.plot(xs,ps,label='Bad Poly fit')
plt.legend();
```

```python
#above the polynomial fit is pulled up at low values due to existence of Outliers
x,y=xs.copy(),ys.copy()
for i in range(5):
    p=np.polyfit(x,y,deg=5)
    ps = np.polyval(p,x)
    Good=y-ps < 3# Mask : Only remove positive outliers
    x_bad,y_bad=x[~Good],y[~Good]
    x,y=x[Good],y[Good]

    plt.plot(x,y,".",label='Used Data',alpha=0.25)
    plt.plot(x,np.polyval(p,x),label=f'Poly fit{i}')
    plt.plot(x_bad,y_bad,".",label='Not used Data')
    plt.legend()
    plt.show()
    
    if (~Good).sum()==0 :
        break
```

Much better way to do the above :i.e removal of Outlier

```python
#use smart alogorithm sklearn: Automating it:Its Machine Learning tool
from sklearn.neighbors import LocalOutlierFactor

lof =LocalOutlierFactor(n_neighbors=20,contamination=0.01)
good = lof.fit_predict(d2) == 1
plt.scatter(d2[good,0],d2[good,1],label='Good',color='#4CAF50')
plt.scatter(d2[~good,0],d2[~good,1],label='Bad',color='#F44336')
plt.legend();
```
