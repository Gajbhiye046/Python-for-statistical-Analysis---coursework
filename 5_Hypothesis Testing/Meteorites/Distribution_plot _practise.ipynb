```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs =np.linspace(-5,10,200)
ks=np.arange(50)
```

```python
ks
```

### Discrete pmf(Probability Mass function)
Binomial distribution, Poissons distribution

```python
pmf_binom=st.binom.pmf(ks,50,0.25)
plt.bar(ks,pmf_binom,label='Binomial example(dice)',alpha =0.8)

pmf_poisson=st.poisson.pmf(ks,30)
plt.bar(ks,pmf_poisson,label='Car crash',alpha =0.8)
plt.legend()

# for particular values
print(st.binom.pmf(10,50,0.25))
print(st.poisson.pmf(50,30))
```

### Continuous distributions
Uniform,normal,exponential , Log normal, skew normal


```python
pdf_normal= st.norm.pdf(xs,loc=5,scale=2)
plt.plot(xs,pdf_normal , label= 'Normal(5,2)');

pdf_exponential= st.expon.pdf(xs,loc=-2,scale=2)
plt.plot(xs,pdf_exponential , label= 'Exponential(5,2)');

pdf_studentt= st.t.pdf(xs,1)
plt.plot(xs,pdf_studentt , label= 'student-t(1)');

pdf_lognormal= st.lognorm.pdf(xs,1)
plt.plot(xs,pdf_lognormal , label= 'Log normal(1)');

pdf_skewnorm= st.skewnorm.pdf(xs,-6)
plt.plot(xs,pdf_skewnorm , label= 'Skewnorm(-6)');

plt.legend()
plt.xlabel('x')
plt.ylabel('Probability');
```
